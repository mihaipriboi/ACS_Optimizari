```julia
Pkg.add("JuMP")
Pkg.add("HiGHS")
using JuMP
using HiGHS
```

```julia
IC = Model(HiGHS.Optimizer)

@variable(IC, xA >= 0)
@variable(IC, xB >= 0)

@objective(IC, Max, 4*xA + 6*xB)

@constraint(IC, 2*xA <= 14)
@constraint(IC, 3*xB <= 15)
@constraint(IC, 4*xA + 3*xB <= 36)

print(IC)
optimize!(IC)

println("Status: $(termination_status(IC))")

if termination_status(IC) == MOI.OPTIMAL
  println("Valoare functie obiectiv: $(objective_value(IC))")
  println("xA: ", value(xA))
  println("xB: ", value(xB))
else
  println("Nu exista solutii disponibile")
end
```

```julia
IC = Model(HiGHS.Optimizer)

@variable(IC, xA >= 0, Int)
@variable(IC, xB >= 0, Int)

@objective(IC, Max, 4*xA + 6*xB)

@constraint(IC, 2*xA <= 14)
@constraint(IC, 3*xB <= 15)
@constraint(IC, 4*xA + 3*xB <= 36)

print(IC)
optimize!(IC)

println("Status: $(termination_status(IC))")

if termination_status(IC) == MOI.OPTIMAL
  println("Valuare functie obiectiv: $(objective_value(IC))")
  println("xA: ", round(Int64, value(xA)))
  println("xB: ", round(Int64, value(xB)))
else
  println("Nu sunt disponibile solutii")
end
```

```julia
Scaune = ["A", "B", "C", "D", "E", "F", "G", "H", "I", "J"]
C = length(Scaune)

LiniiProductie = [1, 2, 3, 4, 5]
P = length(LiniiProductie)

Profit = [6, 5, 9, 5, 6, 3, 4, 7, 4, 3]

Capacitate = [47, 19, 36, 13, 46]

ResurseFolosite = [
    6 4 2 3 1 10 2 9 3 5;
    5 6 1 1 7 2 9 1 8 6;
    8 10 7 2 9 6 9 6 5 6;
    8 4 8 10 5 4 1 5 3 5;
    1 4 7 2 4 1 2 3 10 1
]
```

```julia
IC2 = Model(HiGHS.Optimizer)

@variable(IC2, x[c=1:C] >= 0)

@objective(IC2, Max, sum(Profit[c] * x[c] for c=1:C))

@constraint(IC2, [p=1:P],
    sum(ResurseFolosite[p,c] * x[c] for c=1:C) <= Capacitate[p]
)

optimize!(IC2)

println("Status final: $(termination_status(IC2))")

if termination_status(IC2) == MOI.OPTIMAL
    println(" Valoarea functiei obiectiv: $(objective_value(IC2))")
    for c = 1:C
        if value(x[c]) > 0.001
            println("Scaun de tip: ", Scaune[c], " a fost produs in catntitatea: ", value(x[c]))
        end
    end
else
    println("Nici o solutie disponibila")
end
```

```julia
Fabrici = ["P1", "P2"]
P = length(Fabrici)

Depozite = ["D1", "D2", "D3", "D4"]
D = length(Depozite)

F = 0.375

Capacitate = [7500, 8500]
Cerere = [3250, 3500, 3500, 3000]

Distante = [
  137  92  48 173;
   54 109 111  85
]
```

```julia
ModelTransport = Model(HiGHS.Optimizer)

@variable(ModelTransport, x[p=1:P, d=1:D] >= 0)

@objective(ModelTransport, Min, sum(F * Distante[p,d] * x[p,d] for p=1:P, d=1:D))

@constraint(ModelTransport, [p=1:P], sum(x[p,d] for d=1:D) <= Capacitate[p])
@constraint(ModelTransport, [d=1:D], sum(x[p,d] for p=1:P) >= Cerere[d])

optimize!(ModelTransport)

println("Status final: $(termination_status(ModelTransport))")

if termination_status(ModelTransport) == MOI.OPTIMAL
  println(" Costul minim de transport este: $(objective_value(ModelTransport))")
  for p = 1:P
    for d = 1:D
      if value(x[p,d]) > 0.001
        println("De la ", Fabrici[p], " la ", Depozite[d], " transportam: ", round(Int, value(x[p,d])))
      end
    end
  end
else
  println("Nici o solutie disponibila")
end
```
